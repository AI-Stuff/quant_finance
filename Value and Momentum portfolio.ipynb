```python
%matplotlib inline
import pandas as pd
import seaborn as sns
import math
```

```python
import pylab
sns.set(font_scale=2)
pylab.rcParams['figure.figsize'] = (16, 12)
```

```python
factors=pd.read_csv('AQR.csv',parse_dates=['DATE'],index_col=0)
```

```python
factors.columns
```

```python
combined=(factors[['VAL','MOM']]/2).sum(axis=1)
```

```python
halflife=36
(pd.Series.rolling(combined,halflife).mean()/pd.Series.rolling(combined,halflife).std())['1980':].plot(title='Rolling Sharpe')
```

# Formula
$$ \frac{\frac{1}{N} \sum r_i}{\sigma_i} $$
