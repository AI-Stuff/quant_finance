```python
%matplotlib inline
import pandas as pd
import statsmodels.api as sm
import numpy as np
```

```python
import seaborn as sns
import pylab
sns.set(font_scale=3)
pylab.rcParams['figure.figsize'] = (24, 18)
```

```python
pd.Series(np.random.randn(300)).plot(kind='hist')
```

```python
ex=pd.Series(np.random.randn(300))
en=pd.Series(np.random.randn(300))
en2=pd.Series(np.random.randn(300))
en3=pd.Series(np.random.randn(300))
```

```python
res=sm.OLS(en3,en+ex+en2+en3+en3).fit(cov_type='HAC',cov_kwds={'maxlags':1})
print(res.summary())
```
